```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.
def sum_of_even(numbers):
    total = 0
    for i in numbers:
        if i % 2 == 0:
            total += i
    return total
x = [1, 2, 3, 4, 5, 6, 7, 78, 88, 9]
print("Sum of even numbers is:", sum_of_even(x))
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200
def check_prime(numbers):
    if numbers > 1:
        for i in range(2, numbers):
            if (numbers % i) == 0:
                return f"{numbers} is not a prime number"
        return f"{numbers} is a prime number"
    else:
        return f"{numbers} is not a prime number"

# Example test
print(check_prime(2))
print(check_prime(15))
print(check_prime(17))

```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0
    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            if self.count == 0:
                self.count += 1
                return self.a
            elif self.count == 1:
                self.count += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration

fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i


# Example usage:
for value in powers_of_two(5):
    print(value, end=" ")
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()   # strip() removes \n at the end


# Example usage:
for line in read_file_line_by_line("sample.txt"):
    print(line)
```
