**Q2. What is an interpreter in Python?**

- An interpreter in Python is a program that reads and runs the Python code line by line. It checks each line one at a time and shows the output. If there is any error, it stops and shows the error.

- Python uses an interpreter instead of a compiler. That is  why we don’t need to compile the whole program before running. It makes testing and debugging easy. The Python interpreter helps in running the code directly.

**Q3. What are pre-defined keywords in Python?**

- Pre-defined keywords in Python are special words that have a fixed meaning. These words are already defined by Python and we cannot use them as variable names.

- Some examples are: if, else, for, while, True, False, def, return, etc.


**Q4. Can keywords be used as variable names? **

No, keywords cannot be used as variable names in Python.

- If we try to use a keyword as a variable name, Python will show an error.

- exaple :  we cannot write if = 5 or while = "hello" because if and while are keywords.


**Q5.What is mutability in Python? **

- Mutability in Python means whether a value can be changed after it is created.
- mutability = mutable + immutable

- If a data type is mutable, it means we can change its content without changing its identity. Example: list, dict, set are mutable types.

- If a data type is immutable, we cannot change its content once it is created. Example: int, float, str, tuple are immutable types.

**Q6. Why are lists mutable, but tuples are immutable?**

- Lists are mutable because we can change, add, or remove items from a list after it is created. For example, we can do list[0] = 10.
Tuples are immutable because once we create a tuple, we cannot change its values. This makes tuples more safe and fast for fixed data.

- The main reason is that Python made lists for dynamic data and tuples for fixed data.


**Q7. What is the difference between “==” and “is” operators in Python?**

The is operator checks if both variables are pointing to the same object in memory.

- Implicit type casting is done automatically by Python when needed. Example:

- Explicit type casting is done by the programmer using functions like int(), float(), str(), etc.

- Conditional statements are used to make decisions in a Python program.

- They help the program decide which block of code to run based on a condition. If the condition is true, one block runs; if false, another block runs.

**Q13. What is the difference between for and while loops?**

- A for loop is used when we know how many times we want to run the loop.
- A while loop is used when we want to keep running until a certain condition becomes false.

```python
#Q 14. Write a program to reverse a string using a while loop
# Reverse a string using while loop

text = input("Enter a string: ")
reversed_text = ""
i = len(text) - 1  # Start from the last index

while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

```python
#Q.15Write a Python program to calculate the factorial of a number provided by the user using a while loop:
number = int(input("enter the number "))
factorial = 1
i = 1
while(i<=number):
  factorial = factorial*i
  i = i+1
print(factorial)
```
